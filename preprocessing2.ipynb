```python
test = []
for num in tqdm(range(len(plans))):
    fea = []
    #處理str 把文字拿掉只取值出來
    ary = plans[num].replace('""','1')
    ary = ary.replace('"','')
    ary = ary.replace('},','')
    ary = ary.replace('}','')
    ary = ary.replace('[','')
    ary = ary.replace(']','')
    ary = ary.replace('distance: ',',')
    ary = ary.replace(' price: ','')
    ary = ary.replace(' eta: ','')
    ary = ary.replace(' transport_mode: ','')
    ary = ary.replace('{','')
    ary = ary.replace(' ','')
    ary = ary.split(',')
    #丟掉ary[0] = '' 這個值 把其餘正確的值append到fea內
    for i in range(len(ary)-1):
        fea.append(int(ary[i+1]))
    #建立正確的值順序以便待會丟入pandas    
    pla = [0]*44 #缺值全0
    #缺值補最大值
#     pla = [0,212163,15000,72992,
#            0,92247,3400,11083,
#            0,206865,0,14529,
#            0,206865,92300,14769,
#            0,5000,0,4708,
#            0,10000,0,3027,
#            0,225864,13100,72638,
#            0,182949,20400,15726,
#            0,90860,3300,10432,
#            0,98894,12800,10433,
#            0,161480,5600,21022]
    
    #將pla 把所有數據放好  之後直接轉pandas
    for i in range(3,len(fea),4):
        
        #有重複的推薦mode則放入第一個重複的推薦
        mode = fea[i]
        if(pla[4*(mode-1)] == 0):  #多加這行
            pla[4*(mode-1)] = 1
            pla[4*(mode-1) +1] = fea[i-3]
            pla[4*(mode-1) +2] = fea[i-2]
            pla[4*(mode-1) +3] = fea[i-1]

    test.append(pla)
    del fea,ary,pla
```

```python
data = pd.DataFrame(test,columns=['mode1', 'dis1', 'price1', 'eta1',
                                'mode2', 'dis2', 'price2', 'eta2',
                                'mode3', 'dis3', 'price3', 'eta3',
                                'mode4', 'dis4', 'price4', 'eta4',
                                'mode5', 'dis5', 'price5', 'eta5',
                                'mode6', 'dis6', 'price6', 'eta6',
                                'mode7', 'dis7', 'price7', 'eta7',
                                'mode8', 'dis8', 'price8', 'eta8',
                                'mode9', 'dis9', 'price9', 'eta9',
                                'mode10', 'dis10', 'price10', 'eta10',
                                'mode11', 'dis11', 'price11', 'eta11'])
```
