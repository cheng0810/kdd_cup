```python
import pandas as pd
import numpy as np
import os
import math
import csv
import random
from tqdm import tqdm_notebook as tqdm
```

```python
test_merges = pd.read_csv('/storage/kdd/test_merges.csv')
test_queries = pd.read_csv('/storage/kdd/test_queries.csv')
```

```python
# test_merges[test_merges.sid == 1126541].plans.values

test_sid = test_queries['sid'].values

transport_mode = []

#put the submission order's test_plan.value to array 
for num in tqdm(range(len(test_sid))):
    transport_mode.append(test_merges[test_merges.sid == test_sid[num]].plans.values)
```

```python
transport_mode_num = []

 
for i in tqdm(range(len(transport_mode))):
    try:
        # only need the first recommend plans.value
        change = str(transport_mode[i]).split('transport_mode')
        #if transport_mode is 10 or 11 so the strlength will be 2
        if(str(change[1])[4] == str(0)):
            transport_mode_num.append(str(10))
        elif(str(change[1])[4] == str(0)):
            transport_mode_num.append(str(11))
        else:
            transport_mode_num.append(str(change[1])[3])
    except:
        #if dont have the plan then put 0 into array
        transport_mode_num.append(0) #0.65
#         transport_mode_num.append(random.randint(1,11)) #0.61
            

print(len(transport_mode_num))
```
